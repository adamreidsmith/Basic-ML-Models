In this project, we will use a linear model to predict the fuel economy of a vehicle given a few other attributes.  The data contains 398 vehicles made between 1970 and 1982 and provides attributes such as weight, acceleration, and number of cylinders.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

First we import the data, which can be found [here](http://archive.ics.uci.edu/ml/datasets/Auto+MPG).

```python
cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 
        'weight', 'acceleration', 'year', 'origin']
df = pd.read_csv('auto-mpg.data', names=cols, sep=' ', skipinitialspace=True, comment='\t')
df.head()
```

Upon inspecting the data types, we see that the horsepower column is not stored as a float.  Trying to convert the column to floats results in a ValueError:

```python
print(df.info(), '\n')

try:
    df.horsepower.astype(float)
except ValueError as err:
    print(err)
```

We see that the horsepower column contains six '?' values:

```python
len(df[df.horsepower == '?'])
```

```python
sns.boxplot(x=df.horsepower[df.horsepower != '?'].astype(float))
```

We can fill these values with the mean or median horsepower.  Since the data appears skewed, the median serves as the best measure of center.

```python
median_hp = df.horsepower[df.horsepower != '?'].astype(float).median()
df.horsepower = df.horsepower.replace('?', median_hp).astype(float)
```

A pair plot is useful for visualizing pairwise relationships between the variables and a KDE plot of the distribution of each variable:

```python
sns.pairplot(df[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']], diag_kind='kde')
```

Some variables appear strongly corrolated and others less so.  Several of the variables seem to have a non-linear relationship with our target variable, 'mpg'.  This is a good indication that certain non-linear models may be a better fit for this data, however we will stick to a linear model for this project.

We can check value counts to get an idea of the distribution of the categorical variables:

```python
df.cylinders.value_counts()/len(df)
```

```python
df.origin.value_counts()/len(df)
```

The 'origin' variable gives one of three countries of origin of the vehicle. 1 corresponds to the USA, 2 to Germany, and 3 to Japan.  We can replace these observations with the country name to make it clear.

```python
df.origin = df.origin.map(lambda x: 'usa' if x == 1 else 'germany' if x == 2 else 'japan')
```

We then need to vectorize the 'origin' vatiable for use in our model, so we split the 'origin' column into three columns, one for each country, via a one-hot encoding with the pandas method `get_dummies`.

```python
df = pd.get_dummies(df, columns=['origin'], prefix='', prefix_sep='')
df.head()
```

Next, we investigate ratios of existing variables and check their correlation with the target variable. If we find new variables with a strong correlation to 'mpg', we can add them to the training dataset as they will aid in producing an accurate model.

```python
dfc = df.copy()

dfc['accel_on_power'] = df.acceleration / df.horsepower
dfc['accel_on_cyl'] = df.acceleration / df.cylinders
dfc['accel_on_disp'] = df.acceleration / df.displacement
dfc['weight_on_power'] = df.weight / df.horsepower
dfc['power_on_disp'] = df.horsepower / df.displacement
dfc['power_on_cyl'] = df.horsepower / df.cylinders

dfc_corr = dfc.corr()
abs(dfc_corr['mpg']).sort_values(ascending=False)
```

We see that 'accel_on_disp', 'accel_on_cyl', and 'accel_on_power' have a stronger correlation with the target variable than some of the original variables, so we will add these new variables to the dataset.

```python
df[['accel_on_power', 'accel_on_cyl', 'accel_on_disp']] = dfc[['accel_on_power', 'accel_on_cyl', 'accel_on_disp']]
df.head()
```

We will preform the regression with the machine learning library PyTorch.

```python
import torch as t
from torch import nn
from torch.optim import SGD
from torch.utils.data import Dataset, DataLoader, random_split
from scipy.stats import linregress
```

We first define a simple class representing a linear model.

```python
class LinearRegression(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearRegression, self).__init__()
        
        self.linear = nn.Linear(input_size, output_size)
    
    def forward(self, x):
        return self.linear(x)

model = LinearRegression(df.shape[1] - 1, 1)
```

The goal of our model is to find parameters $m_1,...,m_n,b$ such that the target variable 'mpg', denoted $y_{mpg}$, is given by  $$y_{mpg} = m_1x_1 + ... + m_nx_n + b$$ where the $x_i$ are our training variables (cylinders, displacement, etc.).


We will run our model for 1000 epochs with a learning rate of 0.01 using a mean squared error loss function.  A simple gradient descent algorithm should suffice to train out model.

```python
lr = 0.01
n_epochs = 1000

loss_func = nn.MSELoss()

optimizer = SGD(model.parameters(), lr=lr)
```

The data is then split into training and testing sets and the independent variables in both sets are standardized using the mean and standard deviation of the training dataset. This is best done in PyTorch by creating a class representing the data, and creating DataLoader objects to iterate over the training and testing datasets.

```python
class Data(Dataset):
    def __init__(self, df):
        self.data = df.drop('mpg', axis=1)
        self.data = (self.data - self.data.mean())/self.data.std()
        self.data = self.data.to_numpy()
        self.mpg = df.mpg
        
    def __len__(self):
        return self.data.shape[0]
    
    def __getitem__(self, i):
        return (self.data[i], self.mpg[i])
```

```python
dataset = Data(df)
train_fraction = 0.5

train_len = int(train_fraction*len(dataset))
train, test = random_split(dataset, (train_len, len(dataset) - train_len))

train_loader = DataLoader(train, batch_size=len(dataset), shuffle=False)
test_loader = DataLoader(test, batch_size=len(dataset), shuffle=False)
```

We set the batch size equal to the length of the full dataset so that the entire training set is used on every parameter update.  For a simple model such as this linear regression model, it is not necessary to split the data into batches.

```python
l = []
model.train()
for epoch in range(n_epochs):
    for x, mpg in train_loader:
        
        optimizer.zero_grad()

        prediction = model(x.type(t.float))

        loss = loss_func(prediction, mpg.unsqueeze(1).type(t.float))
        l.append(loss.detach())

        loss.backward()

        optimizer.step()
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(len(l)), l)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.title('Mean squared error loss vs training epoch')
plt.yscale('log')
```

Now, we can test our model to see how it performs on the testing dataset.

```python
test_data, test_mpg = [*test_loader][0]
```

```python
pred_mpg = model(test_data.type(t.float)).squeeze()

test_mpg = test_mpg.detach().numpy()
pred_mpg = pred_mpg.detach().numpy()
```

Plotting the fuel economy predicted by our model against the reported fuel economy for the test data can give us a good idea of how the model is performing.

```python
plt.figure(figsize=(10,6))
plt.scatter(test_mpg, pred_mpg)
plt.plot(range(10, 46), range(10, 46), 'k--', label='45 degree diagonal')
plt.xlabel('Actual [mpg]')
plt.ylabel('Predicted [mpg]')
plt.title('Actual vs Predicted Fuel Economy')
plt.legend()
```

As seen in the scatter plot above, in the range where actuals are between 10 and 35, the data is scattered symmetrically about the 45 degree diagonal, mostly without large residuals.  This indicates our model is performing well in this range.  For actuals greater than 35, the model appears to underpredict the mpg.  This could be due to the fact that many of our independent variables showed a non-linear correlation with the target variable, however our model assumes a linear relationship. 


Furthermore, the Pearson correlation coefficient computed below indicates a strong correlation between predicted and reported fuel economy, reinforcing that our model is performing well.

```python
regress_params = linregress(test_mpg, pred_mpg)
print('Correlation coefficient:', regress_params.rvalue)
```
